```python
#importing libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv("Mall_Customers.csv") #read the file
```

```python
dataset
```

```python
X = dataset.iloc[:,[3,4]].values #iloc is integer location to select the row and column indices from pandas and dot values are called from pandas and return numpy array.
```

```python
from sklearn.cluster import MeanShift, estimate_bandwidth

bandwidth = estimate_bandwidth(X, quantile=0.2, n_samples=25)
clustModel = MeanShift(bandwidth=bandwidth, bin_seeding=False) 
ms = clustModel.fit_predict(X) #train and test the model
```

```python
ms
```

```python
supervised = dataset #assign dataframe dataset value to supervised variable
```

```python
supervised
```

```python
supervised["Cluster group"] = ms #created a column called "cluster group" in the supervised dataset and assign the predicted values
```

```python
supervised
```

```python
supervised.to_csv("meanshift_cluster.csv", index=False) #created a csv file with supervised data
```

```python
import seaborn as sns

facet = sns.lmplot(data=supervised, x=supervised.columns[3], y=supervised.columns[4], hue=supervised.columns[5],
                   fit_reg=False, legend=True)
```
